```python
def plot2zoomedplot(x,x_zoomed,x_name='e',y_name='final-spin',name_to_save=None,x_wrapper=lambda df:df['e'],y_wrapper=lambda df:df['normal_final-mass']):
    fig, axs = plt.subplots(1,2,figsize=(14,8))
    for i in range(len(x)):
        x_axis=x_wrapper(x[i])
        y_axis=y_wrapper(x[i])
        x_zoomed_axis=x_wrapper(x_zoomed[i])
        y_zoomed_axis=y_wrapper(x_zoomed[i])
        
        mass_label=str(np.unique(round(x[i].loc[:,['initial_m1']],2))[0])
        s_label=str(np.unique(round(x[i].loc[:,['initial-separation']],0))[0])
        plt0=axs[0].scatter(x_axis,y_axis,marker='.',label='m1='+mass_label+',s0='+s_label)
        plt0=axs[1].scatter(x_zoomed_axis,y_zoomed_axis,marker='.',label='m1='+mass_label+',s0='+s_label)
    axs[0].set_xlabel(x_name)
    axs[0].set_ylabel(y_name)
    axs[0].grid()
    axs[0].legend(loc='best')
    axs[1].set_xlabel(x_name)
    axs[1].set_ylabel(y_name)
    axs[1].grid()
    fig.suptitle(x_name+' vs '+y_name+' for different initial parameters')
    if name_to_save: 
        plt.savefig(name_to_save+'.png')
    else:
        plt.show()
    return
```

```python
# def piecewise_function2(e,x0,a0,a1,a2,a3,a4):#,
#     [amplitude,f1,f2,phase,decay,k]=[2.6161e-05,926.153622,-741.2561,-239.0159,9.2364,0.95111]
#     if e<=x0:
#         return amplitude*np.exp(decay*e)*np.cos(f1*e+f2*e**2 + phase)+k
#     else:
#         return np.poly1d([a0,a1,a2,a3,a4])(e)#
# piecewise_function_vec=np.vectorize(piecewise_function2)

def poly2_damped_oscillator(t,amplitude,f1,f2,phase,decay,k):#,amplitude2,frequency2,phase2,decay2
    return amplitude*np.exp(decay*t)*np.cos(f1*t+f2*t**2 + phase)+k

from lmfit import Model

def plot_with_fit(x_zoomed,function, guess,x_axis='e',y_axis='final-chi'):
    
    fig, axs = plt.subplots(1,1,figsize=(14,8))
    
    plt0=axs.scatter(x_zoomed[x_axis], x_zoomed[y_axis],c=x_zoomed['#orbits'],marker='.')
    dmodel=Model(function)
    params = dmodel.make_params(**guess)
    result = dmodel.fit(x_zoomed[y_axis], params, t=x_zoomed[x_axis])
    print(result.best_values)

    mass_label=str(np.unique(round(x_zoomed.loc[:,['initial_m1']],2))[0])
    s_label=str(np.unique(round(x_zoomed.loc[:,['initial-separation']],0))[0])
    axs.plot(x_zoomed[x_axis], function(x_zoomed[x_axis],**result.best_values),ls='dotted',label='m1='+mass_label+',s0='+s_label)

    cb0=fig.colorbar(plt0,label="n orbits in sep "+str(sep0))
    axs.set_xlabel(x_axis)
    axs.set_ylabel(y_axis)
    plt.legend(loc='best')
    plt.show()
    return result.best_values
```

```python
from scipy.optimize import curve_fit
def inverse_pl(x,p,a,b):
    return a/(1-x)**p+b

def double_plot_with_fit(x,x_zoomed,function=inverse_pl, guess=np.array([1,1,1]),x_axis='e',y_axis='normal_final-mass'):
    fig, axs = plt.subplots(1,2,figsize=(14,8))
    param_list=[]
    for i in range(len(x)):
        
        plt0=axs[0].scatter(x[i][x_axis], x[i][y_axis],c=x[i]['#orbits'],marker='.')
        plt0=axs[1].scatter(x_zoomed[i][x_axis], x_zoomed[i][y_axis],c=x_zoomed[i]['#orbits'],marker='.')
        
        params, covs = curve_fit(function, x[i][x_axis], x[i][y_axis],guess)
        param_list.append(params)
        print('Best parameters: ',params)
        mass_label=str(np.unique(round(x[i].loc[:,['initial_m1']],2))[0])
        s_label=str(np.unique(round(x[i].loc[:,['initial-separation']],0))[0])
        axs[0].plot(x[i][x_axis], function(x[i][x_axis],*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        axs[1].plot(x_zoomed[i][x_axis], function(x_zoomed[i][x_axis],*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        
    cb0=fig.colorbar(plt0,label="n orbits in sep "+str(sep0))
    axs[0].set_xlabel(x_axis)
    axs[0].set_ylabel(y_axis)
    axs[1].set_xlabel(x_axis)
    axs[1].set_ylabel(y_axis)
    plt.legend(loc='best')
    plt.show()
    return param_list

def double_plot_with_fit_custom_column(x,x_zoomed,x_name='e',y_name='no. orbits',function=inverse_pl, guess=np.array([1,1,1]),x_axis_wrapper=lambda df: df['e'], custom_column_function=lambda df:df['#orbits']):
    fig, axs = plt.subplots(1,2,figsize=(14,8))
    param_list=[]
    for i in range(len(x)):
        
        x_axis=x_axis_wrapper(x[i])
        x_zoomed_axis=x_axis_wrapper(x_zoomed[i])
        y_axis=custom_column_function(x[i])
        y_zoomed_axis=custom_column_function(x_zoomed[i])
        
        plt0=axs[0].scatter(x_axis,y_axis ,c=x[i]['#orbits'],marker='.')
        plt0=axs[1].scatter(x_zoomed_axis, y_zoomed_axis,c=x_zoomed[i]['#orbits'],marker='.')
        
        params, covs = curve_fit(function, x_axis, y_axis,guess)
        param_list.append(params)
        print('Best parameters: ',params)
        mass_label=str(np.unique(round(x[i].loc[:,['initial_m1']],2))[0])
        s_label=str(np.unique(round(x[i].loc[:,['initial-separation']],0))[0])
        axs[0].plot(x_axis, function(x_axis,*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        axs[1].plot(x_zoomed_axis, function(x_zoomed_axis,*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        
    cb0=fig.colorbar(plt0,label="n orbits in sep "+str(sep0))
    axs[0].set_xlabel(x_name)
    axs[0].set_ylabel(y_name)
    axs[1].set_xlabel(x_name)
    axs[1].set_ylabel(y_name)
    plt.legend(loc='best')
    plt.show()
    return param_list
```

# Generation of new dataset with fitted functions

```python
def poly5(x,a0,a1,a2,a3,a4,a5):
    p=np.poly1d([a0,a1,a2,a3,a4,a5])
    return p(x)

def double_damped_oscillator(t,amplitude,f1,f2,phase,decay,k):#,amplitude2,frequency2,phase2,decay2
    return amplitude*np.exp(decay*t)*np.cos(f1*t+f2*t**2 + phase)+k

e_range_s25_dense=np.linspace(0,0.56,560)
e_range_s25=np.linspace(0,0.56,56)
e_range_s11=np.linspace(0,0.26,260)
ones_s11=np.ones(len(e_range_s11))

masses_s11=[0.2,0.33,0.43,0.5]
masses_s25=[0.12,0.14,0.17,0.25,0.29,0.38,0.41,0.44,0.5]

new_dataset_s11=pd.DataFrame([])
new_dataset_s25=pd.DataFrame([])

for i in range(len(best_values_s11)):
    n_orbits_s11=poly5(e_range_s11,*best_params_e_vs_orbits_s11[i])
    finalchi_s11=double_damped_oscillator(e_range_s11,**best_values_s11[i])
    d_s11={'e':e_range_s11,'#orbits':n_orbits_s11,'initial_m1':masses_s11[i]*ones_s11,'initial-separation':11.*ones_s11,'final-chi':finalchi_s11}
    df_s11 = pd.DataFrame(data=d_s11)
    new_dataset_s11=pd.concat([new_dataset_s11, df_s11])

for i in range(len(best_values_s25)):
    e_range=e_range_s25 if masses_s25[i]!=0.5 else e_range_s25_dense
    ones_s25=np.ones(len(e_range))
    
    n_orbits_s25=poly5(e_range,*best_params_e_vs_orbits_s25[i])
    finalchi_s25=double_damped_oscillator(e_range,**best_values_s25[i])
    d_s25={'e':e_range,'#orbits':n_orbits_s25,'initial_m1':masses_s25[i]*ones_s25,'initial-separation':24.6*ones_s25,'final-chi':finalchi_s25}
    df_s25 = pd.DataFrame(data=d_s25)
    new_dataset_s25=pd.concat([new_dataset_s25,df_s25])

new_dataset=pd.concat([new_dataset_s11,new_dataset_s25])
```
