```python
def plot2zoomedplot(x,x_zoomed,x_name='e',y_name='normalized final mass',name_to_save=None,x_wrapper=lambda df:df['e'],y_wrapper=lambda df:df['normal_final-mass']):
    fig, axs = plt.subplots(1,2,figsize=(14,8))
    for i in range(len(x)):
        x_axis=x_wrapper(x[i])
        y_axis=y_wrapper(x[i])
        x_zoomed_axis=x_wrapper(x_zoomed[i])
        y_zoomed_axis=y_wrapper(x_zoomed[i])
        
        mass_label=str(np.unique(round(x[i].loc[:,['q']],2))[0])
        s_label=str(np.unique(round(x[i].loc[:,['initial-separation']],0))[0])
        plt0=axs[0].scatter(x_axis,y_axis,marker='.',label='m1='+mass_label+',s0='+s_label)
        plt0=axs[1].scatter(x_zoomed_axis,y_zoomed_axis,marker='.',label='m1='+mass_label+',s0='+s_label)
    axs[0].set_xlabel(x_name)
    axs[0].set_ylabel(y_name)
    axs[0].grid()
    axs[0].legend(loc='best')
    axs[1].set_xlabel(x_name)
    axs[1].set_ylabel(y_name)
    axs[1].grid()
    fig.suptitle(x_name+' vs '+y_name+' for different initial parameters')
    if name_to_save: 
        plt.savefig(name_to_save+'.png')
    else:
        plt.show()
    return
```

## Transformation properties

```python
from scipy.optimize import curve_fit

class fitClass:
    def __init__(self):
        pass
    
    def q_transformation_parameters(self,t,a,b,c,d):
        amplitude=self.params_to_take['amplitude']
        f1=self.params_to_take['f1']
        f2=self.params_to_take['f2']
        phase=self.params_to_take['phase']
        decay=self.params_to_take['decay']
        k=self.params_to_take['k']
        return poly2_damped_oscillator(t,amplitude*a,b*f1,c*f2,phase+d,decay,k)-self.traslation

    def s0_transformation_parameters(self,t,a,b,c,d,e,k2):
        amplitude=self.params_to_take['amplitude']
        f1=self.params_to_take['f1']
        f2=self.params_to_take['f2']
        phase=self.params_to_take['phase']
        decay=self.params_to_take['decay']
        k=self.params_to_take['k']
        return poly2_damped_oscillator(t,amplitude*a,b*f1,c*f2,phase+d,decay*e,k)+k2
    
    def s0_transformation_parameters_decay(self,t,a,b,c,d,k2):
        amplitude=self.params_to_take['amplitude']
        f1=self.params_to_take['f1']
        f2=self.params_to_take['f2']
        phase=self.params_to_take['phase']
        decay=self.params_to_take['decay']
        k=self.params_to_take['k']
        return poly2_damped_oscillator(t,amplitude*a,b*f1,c*f2,phase+d,decay,k)+k2
    
    def s0_transformation_parameters_phase(self,t,a,b,c,d,k2):
        amplitude=self.params_to_take['amplitude']
        f1=self.params_to_take['f1']
        f2=self.params_to_take['f2']
        phase=self.params_to_take['phase']
        decay=self.params_to_take['decay']
        k=self.params_to_take['k']
        return poly2_damped_oscillator(t,amplitude*a,b*f1,c*f2,phase,decay*d,k)+k2
    
    def s0_transformation_parameters_f2(self,t,a,b,c,d,k2):
        amplitude=self.params_to_take['amplitude']
        f1=self.params_to_take['f1']
        f2=self.params_to_take['f2']
        phase=self.params_to_take['phase']
        decay=self.params_to_take['decay']
        k=self.params_to_take['k']
        return poly2_damped_oscillator(t,amplitude*a,b*f1,f2,c*phase,decay*d,k)+k2
    
    def s0_transformation_parameters_amp(self,t,a,b,c,d,k2):
        amplitude=self.params_to_take['amplitude']
        f1=self.params_to_take['f1']
        f2=self.params_to_take['f2']
        phase=self.params_to_take['phase']
        decay=self.params_to_take['decay']
        k=self.params_to_take['k']
        return poly2_damped_oscillator(t,amplitude,a*f1,b*f2,c*phase,decay*d,k)+k2
```

## Normalized with orbit number

```python
from scipy.optimize import curve_fit
def inverse_pl(x,p,a,b):
    return a/(1-x)**p+b

def double_plot_with_fit(x,x_zoomed,function=inverse_pl, guess=np.array([1,1,1]),x_axis='e',y_axis='normal_final-mass'):
    fig, axs = plt.subplots(1,2,figsize=(14,8))
    param_list=[]
    for i in range(len(x)):
        
        plt0=axs[0].scatter(x[i][x_axis], x[i][y_axis],c=x[i]['#orbits'],marker='.')
        plt0=axs[1].scatter(x_zoomed[i][x_axis], x_zoomed[i][y_axis],c=x_zoomed[i]['#orbits'],marker='.')
        
        params, covs = curve_fit(function, x[i][x_axis], x[i][y_axis],guess)
        param_list.append(params)
        print('Best parameters: ',params)
        mass_label=str(np.unique(round(x[i].loc[:,['q']],2))[0])
        s_label=str(np.unique(round(x[i].loc[:,['initial-separation']],0))[0])
        axs[0].plot(x[i][x_axis], function(x[i][x_axis],*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        axs[1].plot(x_zoomed[i][x_axis], function(x_zoomed[i][x_axis],*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        
    cb0=fig.colorbar(plt0,label="n orbits in sep "+str(sep0))
    axs[0].set_xlabel(x_axis)
    axs[0].set_ylabel(y_axis)
    axs[1].set_xlabel(x_axis)
    axs[1].set_ylabel(y_axis)
    plt.legend(loc='best')
    plt.show()
    return param_list

def normalization1(df):
    return df['final-mass']

def double_plot_with_fit_custom_column(x,x_zoomed,function=inverse_pl, guess=np.array([1,1,1]),x_axis_wrapper=lambda df: df['e'], custom_column_function=normalization1):
    fig, axs = plt.subplots(1,2,figsize=(14,8))
    param_list=[]
    for i in range(len(x)):
        
        x_axis=x_axis_wrapper(x[i])
        x_zoomed_axis=x_axis_wrapper(x_zoomed[i])
        y_axis=custom_column_function(x[i])
        y_zoomed_axis=custom_column_function(x_zoomed[i])
        
        plt0=axs[0].scatter(x_axis,y_axis ,c=x[i]['#orbits'],marker='.')
        plt0=axs[1].scatter(x_zoomed_axis, y_zoomed_axis,c=x_zoomed[i]['#orbits'],marker='.')
        
        params, covs = curve_fit(function, x_axis, y_axis,guess)
        param_list.append(params)
        print('Best parameters: ',params)
        mass_label=str(np.unique(round(x[i].loc[:,['q']],2))[0])
        s_label=str(np.unique(round(x[i].loc[:,['initial-separation']],0))[0])
        axs[0].plot(x_axis, function(x_axis,*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        axs[1].plot(x_zoomed_axis, function(x_zoomed_axis,*params),ls='dotted',label='m1='+mass_label+',s0='+s_label)
        
    cb0=fig.colorbar(plt0,label="n orbits in sep "+str(sep0))
    axs[0].set_xlabel('alternative e')
    axs[0].set_ylabel('alternative final mass')
    axs[1].set_xlabel('alternative e')
    axs[1].set_ylabel('alternative final mass')
    plt.legend(loc='best')
    plt.show()
    return param_list
```
